```python
import pandas as pd
```

```python
df_1 = pd.read_csv("data/athlete_events.csv")
df_2 = pd.read_csv("data/noc_regions.csv")
df_3 = pd.read_csv("data/countries.csv")
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_3.head()
```

```python
merged_table_1 = pd.merge(df_1, df_2, on="NOC",  how="outer")
del merged_table_1['notes']
dataframe_1 = merged_table_1.rename(columns={"region":"name"})
dataframe_1.head()
```

```python
bool_series = pd.isnull(dataframe_1["Games"])
dataframe_1[bool_series]
```

```python
cleaned = dataframe_1.drop(271116, axis=0)
cleaned.head()
```

```python
olympic_data = pd.merge(cleaned, df_3, on="name")
olympic_data.head()
```

```python
olympicDataFinal = olympic_data[["name","latitude","longitude","Medal","Games"]]
olympicDataFinal.head()
```

```python
from sqlalchemy import create_engine
```

```python
# Create the connection engine
engine = create_engine("sqlite:///db/olympicDataFinal.sqlite")
conn = engine.connect()
```

```python
olympicDataFinal.to_sql("olympic_data", conn, index=False, if_exists="replace")
```
